# RGB to HSV conversion

```python
def f_rgb_to_hsv(r, g, b,scaleFactor): 
    r, g, b = r / 255.0, g / 255.0, b / 255.0
    cmax = max(r, g, b)    # maximum of r, g, b 
    cmin = min(r, g, b)    # minimum of r, g, b 
    diff = cmax-cmin       # diff of cmax and cmin. 
    if cmax == cmin:  
        h = 0
    elif cmax == r:  
        h = (60 * ((g - b) / diff) + 0) % 360
    elif cmax == g: 
        h = (60 * ((b - r) / diff) + 120) % 360
    elif cmax == b: 
        h = (60 * ((r - g) / diff) + 240) % 360
   
    if h < 0:
        h = h + 360
    if cmax == 0: 
        s = 0
    else: 
        s = (diff / cmax) * scaleFactor
    v = cmax * scaleFactor
    return h, s, v 
```

```python
print(f_rgb_to_hsv(100, 200, 50, 100))
```

```python
tulips = cv2.imread(r'C:\Users\User\Desktop\Python_projects\Computer_vision_lessons_and_exercises\tulips.jpg')
```

```python
plt.imshow(tulips[:, :, ::-1])
```

```python
hsv_tulips = cv2.cvtColor(tulips, cv2.COLOR_BGR2HSV)
```

```python
# here we will show the red flowers only
# so we take the levels of red 
lower_red = np.array([0, 120, 70])      # 0 for HUE, 120 for SATURATION and 70 to BRIGHTNESS
upper_red = np.array([15, 255, 255])
mask1 = cv2.inRange(hsv_tulips, lower_red, upper_red)
```

```python
print(mask1.shape, mask1.dtype, mask1.max(), mask1.min(), sep='\n')
```

```python
lower_red = np.array([165, 120, 70])      # 160 for HUE, 120 for SATURATION and 70 to BRIGHTNESS
upper_red = np.array([180, 255, 255])
mask2 = cv2.inRange(hsv_tulips, lower_red, upper_red)
```

```python
np.unique(mask2)
```

```python
mask = mask1 | mask2
np.unique(mask)
```

```python
result = cv2.bitwise_and(tulips, tulips, mask=mask)
```

```python
# show original
plt.figure()
plt.subplot(121)
plt.imshow(tulips[:, :, ::-1])

# show result
# if not accurate enough, change lower and upper red values
plt.subplot(122)
plt.imshow(result[:, :, ::-1])
plt.show()
```
