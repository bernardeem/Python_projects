# Introduction to geometric transformations

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
color_image = r'C:\Users\User\Desktop\Python_projects\Computer_vision_lessons_and_projects\tulips.jpg'
bgr_img = cv2.imread(color_image, cv2.IMREAD_COLOR)
bgr_img = bgr_img[:, :, ::-1]
plt.imshow(bgr_img)
```

```python
bgr_img_resized = cv2.resize(src=bgr_img, fx=0.5, fy=2, dsize=None)
plt.imshow(bgr_img_resized[:, :, ::-1])
```

```python
gray_image_path = r'C:\Users\User\Desktop\Python_projects\Computer_vision_lessons_and_projects\albert-einstein_gray.jpg'
img_gray = cv2.imread(gray_image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# resize image to double it horizontally and vertically it will be reduced to half of the original size
img_gray_resized = cv2.resize(src=img_gray, fx=1.5, fy=0.8, dsize=None)

plt.imshow(img_gray_resized, cmap='gray')
plt.xticks([])
plt.yticks([])
```

```python
print(img_gray_resized.shape, img_gray.shape)
```
