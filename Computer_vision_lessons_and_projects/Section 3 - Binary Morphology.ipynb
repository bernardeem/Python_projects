# Binary Morphology

Similar to a grayscale image, the binary image has values 0 [black] or 1 [white]. 
Which corresponds to grayscale black = 0 and to grayscale white = 255.

## Erosion
The process of image erosion involves shrinking the foreground regions by removing pixels from their boundaries.
Exemple:
image = cv2.imread('your_image.png', 0)
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(image, kernel)

## Dilation
Used to enhance or expand regions of interest in a binary image.
Example:
image = cv2.imread('your_image.png', 0)
kernel = np.ones((5, 5), np.uint8)
dilated_image = cv2.dilate(image, kernel)

## Opening
The proccess of erosion followed by dilation

## Closing
The reverse process of opening

## Gradient
Operation that computes the difference between the dilation and erosion of a binary image
Highlight the boundaries of the image

## Transforming an image into a binary image

```python
image = plt.imread(r'C:\Users\User\Desktop\Python_projects\Computer_vision_lessons_and_projects\MIT_example.png')
print(image.dtype)
if len(np.unique(image)) != 2:
    print('The image IS NOT binary')
else:
    print('The image IS binary')

plt.imshow(image, cmap='gray')
```

```python
# the values are between 0 an 1
threshold = 0.6
im = image > threshold
```

```python
# transforming so it can be read and shown by plt
im_uint8 = (im * 255).astype(np.uint8)

print(np.unique(im_uint8))
plt.imshow(im_uint8, cmap='gray')
```
