# Binary Morphology

Similar to a grayscale image, the binary image has values 0 [black] or 1 [white]. 
Which corresponds to grayscale black = 0 and to grayscale white = 255.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = plt.imread(r"C:\Users\User\Desktop\Python_projects\Computer_vision_lessons_and_projects\albert-einstein_gray.jpg")
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
threshold = 10
im2 = img_gray > threshold    # if bigger than 100, True or 1. Else is False or 0.
```

```python
plt.imshow(im2, cmap='gray')
```

```python
plt.imshow(~(im2), cmap='gray')   # when using the operator '~', you can see the opposite
print(np.unique(im2))
```

```python
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
print(type(kernel_rect))
print(kernel_rect.dtype)
print(kernel_rect)
```
