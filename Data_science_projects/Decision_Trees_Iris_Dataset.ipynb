### Decision Tree project using Iris dataset

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
import numpy as np
```

```python
df = load_iris()
X = df.data[:, 2 :]
y = df.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
export_graphviz(
    tree_clf,
    out_file='iris_tree.dot',
    feature_names=df.feature_names[2:],
    class_names= df.target_names,
    filled=True
)

graph = pydotplus.graph_from_dot_file('iris_tree.dot')
graph.write_png('iris_tree.png')

Image('iris_tree.png')
```

```python
# estimating class probability
sample = np.array([[5, 1.5]])
class_probabilities = tree_clf.predict_proba(sample)
predicted_class = tree_clf.predict(sample)
print('Class probabilities:', class_probabilities)
print('Predicted class:', predicted_class)
```

#### Applying Decision Tree Regression
Decision trees are capable of doing regression tasks

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=3)
tree_reg.fit(X, y)

export_graphviz(
    tree_reg,
    out_file='iris_tree_regression.dot',
    feature_names=df.feature_names[2:],
    class_names= df.target_names,
    filled=True
)

graph = pydotplus.graph_from_dot_file('iris_tree_regression.dot')
graph.write_png('iris_tree_regression.png')

Image('iris_tree_regression.png')
```
