# Simple Polynomial Regression example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mantain the same random numbers for reproducibility
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * (X**2) + X + 2 + np.random.randn(m, 1)
```

```python
plt.plot(X, y, 'o', markersize=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dot Plot of the randomly generated Data')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
print(X[0])
print(X_poly[0])   # x and x**2
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_poly, y)
lr.intercept_, lr.coef_
```

It estimates y = 0.564*(x^2) + 0.933*x + 1.78

```python
X_new=np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lr.predict(X_new_poly)

plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10])
plt.show()
```

This approach is extremely useful in case of non-linear data.
