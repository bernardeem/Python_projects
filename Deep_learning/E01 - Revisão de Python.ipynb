## 7

Utilizando as classes acima, crie uma lista com $20$ objetos de cada classe (total de $60$ objetos), todos com propriedades, tais como `lado`, `base`, `altura` e `raio` aleatórias, todas com valores sorteados no intervalo entre `1` e `10`. Para fazer isso, utilize laço `for` (criando $20$ objetos de cada classe) e para criar as propriedades aleatórias use a função `randint(1,10)` do submódulo `random` do `numpy` (por exemplo, use `from numpy.random import randint` no início do código). Coloque todos objetos em uma única lista (não faça uma lista de listas, mas sim uma única lista, que vai sendo preenchida em cada laço com os $60$ objetos). O nome da variável que deve conter essa lista final deve ser `objs`.


```python
# Seu código aqui
from numpy.random import randint

class Quadrado:
  lado = 0
  def __init__(self, l):    # para passar o valor de lado já ao chamar q1 = Quadarado(5)
    self.lado = l

  def area(self):             # é um método
    return self.lado**2

q1 = Quadrado(3)
print(q1.area())

class Triangulo:
  base = 0
  altura = 0
  def __init__(self, b, h):
    self.base = b
    self.altura = h
  def area(self):
    return self.base * (self.altura / 2)

t1 = Triangulo(4, 5)
print(t1.area())


class Circulo:
  raio = 0
  def __init__(self, r):
    self.raio = r
  def area(self):
    return np.pi * (self.raio**2)

c1 = Circulo(10)
print(c1.area())


objs = []
soma = 0

for i in range(60):
  if i < 20:
    objeto = 0
    objeto = Quadrado(randint(1, 10))
    objs.append(objeto)
    print('Área do quadrado gerado: ', objeto.area())
    soma += objeto.area()

  elif i < 40:
    objeto = 0
    objeto = Triangulo(randint(1, 10), randint(1, 10))
    objs.append(objeto)
    print('Área do triângulo gerado: ', objeto.area())
    soma += objeto.area()

  else:
    objeto = 0
    objeto = Circulo(randint(1, 10))
    objs.append(objeto)
    print('Área do círculo gerado: ',objeto.area())
    soma += objeto.area()

print(f'A soma das áreas é: {soma:.4f}')
print()
print(len(objs), objs)
```

## 8
Use a função `shuffle()` para embaralhar a ordem dos elementos na lista. A função `shuffle` também está no submódulo `random` do módulo `numpy`, ou seja, precisa ser importada.

Após tudo isso calcule a soma total de todas as áreas desses objetos que estão contidos na lista, e mostre o resultado.

Salve a área total numa variável de nome `soma`

```python
# Seu código aqui
from numpy.random import randint
from numpy.random import shuffle

class Quadrado:
  lado = 0
  def __init__(self, l):    # para passar o valor de lado já ao chamar q1 = Quadarado(5)
    self.lado = l

  def area(self):             # é um método
    return self.lado**2

q1 = Quadrado(3)
print(q1.area())

class Triangulo:
  base = 0
  altura = 0
  def __init__(self, b, h):
    self.base = b
    self.altura = h
  def area(self):
    return self.base * (self.altura / 2)

t1 = Triangulo(4, 5)
print(t1.area())


class Circulo:
  raio = 0
  def __init__(self, r):
    self.raio = r
  def area(self):
    return np.pi * (self.raio**2)

c1 = Circulo(10)
print(c1.area())


objs = []
soma = 0

for i in range(60):
  if i < 20:
    objeto = 0
    objeto = Quadrado(randint(1, 10))
    objs.append(objeto)
    print('Área do quadrado gerado: ', objeto.area())
    soma += objeto.area()

  elif i < 40:
    objeto = 0
    objeto = Triangulo(randint(1, 10), randint(1, 10))
    objs.append(objeto)
    print('Área do triângulo gerado: ', objeto.area())
    soma += objeto.area()

  else:
    objeto = 0
    objeto = Circulo(randint(1, 10))
    objs.append(objeto)
    print('Área do círculo gerado: ', objeto.area())
    soma += objeto.area()

print(f'A soma das áreas é: {soma:.4f}')
print()
print(len(objs), objs)


shuffle(objs)   
print(objs)

#nova_lista = shuffle(objs)
#print(objs[0])


# Verificação
soma_verificada = sum([obj.area() for obj in objs])
if abs(soma_verificada - soma) < 1e-6:
    print("A soma verificada é igual à soma calculada.")
else:
    print("A soma verificada é diferente da soma calculada.")
```
