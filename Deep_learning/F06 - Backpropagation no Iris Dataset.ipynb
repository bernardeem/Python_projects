```python
# Fazendo rede

def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))

class Perceptron:
  def __init__(self):
    # criando matriz de pesos com valores de 0 a 1 da PRIMEIRA CAMADA
    self.Wh1 = np.random.random((5, 4)) * 2.0 - 1.0   # as saídas são as linhas e as entradas são as colunas
                                         # o vezes 2 e menos 1 é pra gerar valores entre -1 e 1
                                         # 5 neurônios na primeira camada
                                         # 4 entradas, sendo elas sl, sw, pl, pw
    # os bias da primeira camada
    self.bh1 = np.random.random((5, 1)) * 2.0 - 1.0        # vetor coluna com 5 linhas de -1 a 1

    # SEGUNDA CAMADA
    self.Wh2 = np.random.random((4, 5)) * 2.0 - 1.0     # 4 neurônios na segunda camada e 5 que são a saída da camada anterior
    # bias da camada de saída
    self.bh2 = np.random.random((4, 1)) * 2.0 - 1.0        # vetor coluna com 4 linhas de -1 a 1

    # CAMADA SAÍDA
    self.Wo = np.random.random((3, 4)) * 2.0 - 1.0     # 3 neurônios na camada de saída e 4 que são a saída da camada anterior
    # bias da camada de saída
    self.bo = np.random.random((3, 1)) * 2.0 - 1.0        # vetor coluna com 3 linhas de -1 a 1

    self.eta = 0.01             # passo de aprendizagem


  def compute(self, x):
    x = np.reshape(x, (4, 1))      # vai transformar em vetor coluna caso venha vetor linha
    # 's' vai ser o resultado da operação linear de soma ponderada do pesos vezes entradas + bias (soma das sinapses vezes resultado
    # da camada anterior vezes os pesos mais os bias --- AINDA ANTES DE PASSAR PELA FUNÇÃO DE ATIVAÇÃO)
    self.sh1 = np.dot(self.Wh1, x) + self.bh1   # multiplicação matricial dos pesos vezes a entrada
    self.zh1 = sigmoid(self.sh1)    # z é a função de ativação

    # CAMADA 2:
    self.sh2 = np.dot(self.Wh2, self.zh1) + self.bh2   # multiplicação matricial dos pesos vezes a entrada
    self.zh2 = sigmoid(self.sh2)    # z é a função de ativação

    # CAMADA DE SAÍDA:
    self.so = np.dot(self.Wo, self.zh2) + self.bo   # saída da camada 1
    self.zo = sigmoid(self.so)
    return self.zo   # retorna a camada de saída

  def backprop(self, X, Y):
      Err = 0.0
      total = X.shape[0]
      for i in range(total):
        x = X[i, :]
        x = np.reshape(x, (4, 1))
        y_hat = Y[i, :]      # famoso y chapéu (^)
        y_hat = np.reshape(y_hat, (3, 1))
        y = self.compute(x)
        err = np.sum((y - y_hat)**2) / 2
        Err += err
        self.do = (y - y_hat) * self.zo - (1 - self.zo)  # basicamente isso: self.zo - (1 - self.zo) É A DERIVADA DA SIGMOIDE
                                                         # NESTE CASO DA SIGMOIDE, APENAS, SE PODE FAZER ISSO!
        self.dh2 = np.dot(self.Wo.T, self.do) * self.zh2 * (1 - self.zh2)     # o '.T' é de matriz transposta
        self.dh1 = np.dot(self.Wh2.T, self.dh2) * self.zh1 * (1 - self.zh1)

        self.Wo -= self.eta * np.dot(self.do, self.zh2.T)
        self.bo -= self.eta * self.do
        self.Wh2 -= self.eta * np.dot(self.dh2, self.zh1.T)
        self.bh2 -= self.eta * self.dh2
        self.Wh1 -= self.eta * np.dot(self.dh1, x.T)
        self.bh1 -= self.eta * self.dh1

      Err /= total    # divide erro pelo total
      return Err
```

```python
import matplotlib.pyplot as plt

p = Perceptron()

for i in range(10000):
  Err = p.backprop(x_train, y_train)
  if not (i % 1000) or i == 0:
    print('Erro: ', Err)

# Plotando o gráfico
plt.plot(Err)
plt.xlabel('Iteração')
plt.ylabel('Erro')
plt.title('Diminuição do Erro ao Longo das Iterações')
plt.show()
```

```python
# Fazendo rede

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

class Perceptron:
    def __init__(self):
        # criando matriz de pesos com valores de 0 a 1 da PRIMEIRA CAMADA
        self.Wh1 = np.random.random((5, 4)) * 2.0 - 1.0   # as saídas são as linhas e as entradas são as colunas
                                                         # o vezes 2 e menos 1 é pra gerar valores entre -1 e 1
                                                         # 5 neurônios na primeira camada
                                                         # 4 entradas, sendo elas sl, sw, pl, pw
        # os bias da primeira camada
        self.bh1 = np.random.random((5, 1)) * 2.0 - 1.0        # vetor coluna com 5 linhas de -1 a 1

        # SEGUNDA CAMADA
        self.Wh2 = np.random.random((4, 5)) * 2.0 - 1.0     # 4 neurônios na segunda camada e 5 que são a saída da camada anterior
        # bias da camada de saída
        self.bh2 = np.random.random((4, 1)) * 2.0 - 1.0        # vetor coluna com 4 linhas de -1 a 1

        # CAMADA SAÍDA
        self.Wo = np.random.random((3, 4)) * 2.0 - 1.0     # 3 neurônios na camada de saída e 4 que são a saída da camada anterior
        # bias da camada de saída
        self.bo = np.random.random((3, 1)) * 2.0 - 1.0        # vetor coluna com 3 linhas de -1 a 1

        self.eta = 0.01             # passo de aprendizagem

    def compute(self, x):
        x = np.reshape(x, (4, 1))      # vai transformar em vetor coluna caso venha vetor linha
        # 's' vai ser o resultado da operação linear de soma ponderada do pesos vezes entradas + bias (soma das sinapses vezes resultado
        # da camada anterior vezes os pesos mais os bias --- AINDA ANTES DE PASSAR PELA FUNÇÃO DE ATIVAÇÃO)
        self.sh1 = np.dot(self.Wh1, x) + self.bh1   # multiplicação matricial dos pesos vezes a entrada
        self.zh1 = sigmoid(self.sh1)    # z é a função de ativação

        # CAMADA 2:
        self.sh2 = np.dot(self.Wh2, self.zh1) + self.bh2   # multiplicação matricial dos pesos vezes a entrada
        self.zh2 = sigmoid(self.sh2)    # z é a função de ativação

        # CAMADA DE SAÍDA:
        self.so = np.dot(self.Wo, self.zh2) + self.bo   # saída da camada 1
        self.zo = sigmoid(self.so)
        return self.zo   # retorna a camada de saída

    def backprop(self, X, Y):
        Err = 0.0
        total = X.shape[0]
        for i in range(total):
            x = X[i, :]
            x = np.reshape(x, (4, 1))
            y_hat = Y[i, :]      # famoso y chapéu (^)
            y_hat = np.reshape(y_hat, (3, 1))
            y = self.compute(x)
            err = np.sum((y - y_hat) ** 2) / 2
            Err += err
            self.do = (y - y_hat) * y * (1 - y)  # gradientes da camada de saída
            self.dh2 = np.dot(self.Wo.T, self.do) * self.zh2 * (1 - self.zh2)  # gradientes da segunda camada oculta
            self.dh1 = np.dot(self.Wh2.T, self.dh2) * self.zh1 * (1 - self.zh1)  # gradientes da primeira camada oculta

            self.Wo -= self.eta * np.dot(self.do, self.zh2.T)  # atualização dos pesos da camada de saída
            self.bo -= self.eta * self.do  # atualização dos bias da camada de saída
            self.Wh2 -= self.eta * np.dot(self.dh2, self.zh1.T)  # atualização dos pesos da segunda camada oculta
            self.bh2 -= self.eta * self.dh2  # atualização dos bias da segunda camada oculta
            self.Wh1 -= self.eta * np.dot(self.dh1, x.T)  # atualização dos pesos da primeira camada oculta
            self.bh1 -= self.eta * self.dh1  # atualização dos bias da primeira camada oculta

        Err /= total    # divide erro pelo total
        return Err
```
