```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image = plt.imread(r'C:\Users\User\Desktop\Python_projects\Image_processing_project\image_1.jpeg')
plt.imshow(image)
```

```python
R = image[:, :, 0]
G = image[:, :, 1]
B = image[:, :, 2]
```

```python
plt.figure(1)


plt.subplot(231)    # 2 rows, 3 columns
plt.imshow(image)
plt.xticks([])      # remove image labels
plt.yticks([])

plt.subplot(232)    # changed last number
plt.imshow(image)
plt.xticks([])      
plt.yticks([])

plt.subplot(233)    
plt.imshow(image)
plt.xticks([])      
plt.yticks([])

plt.subplot(234)    
plt.imshow(R, cmap='gray')
plt.xticks([])      
plt.yticks([])
plt.title('Red')

plt.subplot(235)    
plt.imshow(G, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Green')

plt.subplot(236)   
plt.imshow(B, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Blue')


plt.show()
```

```python
image2 = image[:, :, ::-1]       # inverting because cv2 reads the image as BGR
denoised = cv2.fastNlMeansDenoisingColored(image2, None, 10, 10, 7, 21)

plt.imshow(denoised)
```

```python
plt.figure(1)


plt.subplot(231)    # 2 rows, 3 columns
plt.imshow(image2)
plt.xticks([])      # remove image labels
plt.yticks([])

plt.subplot(232)    # changed last number
plt.imshow(image2)
plt.xticks([])      
plt.yticks([])

plt.subplot(233)    
plt.imshow(image2)
plt.xticks([])      
plt.yticks([])

plt.subplot(234)    
plt.imshow(R, cmap='gray')
plt.xticks([])      
plt.yticks([])
plt.title('Red')

plt.subplot(235)    
plt.imshow(G, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Green')

plt.subplot(236)   
plt.imshow(B, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Blue')


plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
erosion = cv2.erode(image2, kernel)
plt.imshow(erosion)
```

```python
plt.figure(1)


plt.subplot(231)    # 2 rows, 3 columns
plt.imshow(erosion)
plt.xticks([])      # remove image labels
plt.yticks([])

plt.subplot(232)    # changed last number
plt.imshow(erosion)
plt.xticks([])      
plt.yticks([])

plt.subplot(233)    
plt.imshow(erosion)
plt.xticks([])      
plt.yticks([])

plt.subplot(234)    
plt.imshow(R, cmap='gray')
plt.xticks([])      
plt.yticks([])
plt.title('Red')

plt.subplot(235)    
plt.imshow(G, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Green')

plt.subplot(236)   
plt.imshow(B, cmap='gray')
plt.xticks([])     
plt.yticks([])
plt.title('Blue')


plt.show()
```
