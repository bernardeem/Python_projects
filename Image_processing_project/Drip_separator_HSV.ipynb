### Drip separator project using K-Means

Verify channel separation using HSV instead of BGR channels.
Verify if this approach is better than the others. 

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# using image 4 because it is not as simple as the first 2 images
# use plt to to keep the image displayed in the notebook
image = cv2.imread('image_4.jpeg')
example_img = image[:, :, -1]
plt.imshow(example_img)
```

```python
# THIS CELL IS EXTREMELY USEFUL, TRY "SATURATION" AND "VALUE"
# CHANNELS TO COUNT THE DRIPS
# Convert BGR to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Separate channels
hue_channel, saturation_channel, value_channel = cv2.split(hsv_image)

# Display original image and separated channels
cv2.imshow('Original Image', image)
cv2.imshow('Hue Channel', hue_channel)
cv2.imshow('Saturation Channel', saturation_channel)
cv2.imshow('Value Channel', value_channel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# try using SATURATION and VALUE to count contours
```

```python
# convert image to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# define range of blue color in HSV
lower_blue = np.array([100,50,50])
upper_blue = np.array([130,255,255])

# threshold the HSV image to get only blue colors
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# find contours
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# draw contours on original image (optional)
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

# count the number of contours (number of drips)
num_drips = len(contours)
print("Number of drips:", num_drips)

# display the original image with contours
cv2.imshow('Original Image with Contours', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
